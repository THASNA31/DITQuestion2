```python
#%% Getting the data
NPI_df = pd.read_csv("C:/Users/tanzi/Documents/DataScience/PartD_Prescriber_PUF_NPI_17.txt", sep='\t')
NPI_df.head()
```

```python
# Question 2
Q2table = pd.pivot_table(NPI_df, values=['total_day_supply','total_claim_count'], index='npi', aggfunc=np.sum)
Q2df = pd.DataFrame(Q2table.to_records())
Q2df['average_prescription_length'] = Q2df['total_day_supply']/Q2df['total_claim_count']
print(Q2df['average_prescription_length'].median())
sns.distplot(Q2df['average_prescription_length'])
```

```python
# Question 3
Q3table = pd.pivot_table(NPI_df, values=['total_claim_count','brand_claim_count'], index='specialty_description', aggfunc=np.sum)
Q3df = pd.DataFrame(Q3table.to_records())
Q3df['fraction_of_brand_drug_claims'] = Q3df['brand_claim_count']/Q3df['total_claim_count']
Q3df = Q3df[Q3df.fraction_of_brand_drug_claims != 0]
print(Q3df['fraction_of_brand_drug_claims'].std())
sns.distplot(Q3df['fraction_of_brand_drug_claims'])
```

```python
# Question 4
Q4table = pd.pivot_table(NPI_df, values=['opioid_bene_count','antibiotic_bene_count'], index='nppes_provider_state', aggfunc=np.sum)
Q4df = pd.DataFrame(Q4table.to_records())
Q4df['opiod_to_antibiotic'] = Q4df['opioid_bene_count']/Q4df['antibiotic_bene_count']
Q4df['opiod_to_antibiotic'].max() - Q4df['opiod_to_antibiotic'].min()
```

```python
# Question 5
Q5table = pd.pivot_table(NPI_df, values=['total_claim_count','total_claim_count_ge65','lis_claim_count'], index='npi', aggfunc=np.sum)
Q5df = pd.DataFrame(Q5table.to_records())
Q5df = Q5df[Q5df.lis_claim_count != 0]
Q5df['lis_to_total'] = Q5df['lis_claim_count']/Q5df['total_claim_count']
Q5df['ge65_to_total'] = Q5df['total_claim_count_ge65']/Q5df['total_claim_count']
lis_ge65_corr = Q5df['lis_to_total'].corr(Q5df['ge65_to_total'])
print(lis_ge65_corr)
```

```python
# Question 6
df1 = NPI_df.groupby(['nppes_provider_state','specialty_description']).count()
Q6df = pd.DataFrame(df1.to_records())
#Q6df = Q6df[Q6df['nppes_provider_state','specialty_description','npi']]
Q6df = Q6df.loc[0:, ['nppes_provider_state','specialty_description','npi']]
Q6df.head()
```

```python
# Question 6 continues
df2 = NPI_df.groupby(['nppes_provider_state','specialty_description'])['opioid_claim_count','opioid_day_supply'].mean()
Q6df_1 = pd.DataFrame(df2.to_records())
Q6df_1.head()
```

```python
# Question 6 continues
result = pd.merge(Q6df, Q6df_1, how='left', on=['nppes_provider_state', 'specialty_description']).reset_index(drop = True)
result = result[result.npi > 100].reset_index(drop = True)
result['Average'] = result['opioid_day_supply']/result['opioid_claim_count']
result.head()
```

```python
# Question 6 continues
ratio = result.groupby('specialty_description')['Average'].sum()
ratio = pd.DataFrame(ratio)
ratio = ratio.reset_index()
ratio['Range'] = ratio['Average']/ratio['Average'].sum()
print(ratio['Range'].max())
```

```python
# Getting 2016 data
NPI16_df = pd.read_csv("C:/Users/tanzi/Documents/DataScience/PartD_Prescriber_PUF_NPI_16.txt", sep='\t')
NPI16_df.head()
```

```python
# Question 7
NPI16Cost_df = NPI16_df.loc[0:,['npi','total_drug_cost','total_day_supply']]
NPI16Cost_df['Average_cost_per_day_2016'] = NPI16Cost_df['total_drug_cost']/NPI16Cost_df['total_day_supply']
NPI16Cost = pd.DataFrame(NPI16Cost_df.groupby('npi').mean().to_records())
NPI16Cost = NPI16Cost.drop(columns=['total_drug_cost', 'total_day_supply'])
NPI16Cost.head()
```

```python
NPI17Cost_df = NPI_df.loc[0:,['npi','total_drug_cost','total_day_supply']]
NPI17Cost_df['Average_cost_per_day_2017'] = NPI17Cost_df['total_drug_cost']/NPI17Cost_df['total_day_supply']
NPI17Cost = pd.DataFrame(NPI17Cost_df.groupby('npi').mean().to_records())
NPI17Cost = NPI17Cost.drop(columns=['total_drug_cost', 'total_day_supply'])
NPI17Cost.head()
```

```python
Cost1617 = pd.merge(NPI16Cost, NPI17Cost, how='left', on=['npi']).reset_index(drop = True)
Cost1617['Inflation'] = Cost1617['Average_cost_per_day_2017']/Cost1617['Average_cost_per_day_2016']
Cost1617['Inflation'].mean()
```

```python
# Question 8
Special16 = pd.DataFrame(NPI16_df.groupby(['specialty_description'])['npi'].count()).reset_index()
Special16 = Special16[Special16['npi'] > 1000].reset_index(drop = True)
Special16 = Special16.sort_values(by = 'npi', ascending = False).reset_index(drop = True)
Special16.rename(columns={'npi': 'npi16'}, inplace=True)
Special16.head()
```

```python
Special17 = pd.DataFrame(NPI_df.groupby(['specialty_description'])['npi'].count()).reset_index()
Special17 = Special17[Special17['npi'] > 1000].reset_index(drop = True)
Special17 = Special17.sort_values(by = 'npi', ascending = False).reset_index(drop = True)
Special17.rename(columns={'npi': 'npi17'}, inplace=True)
Special17.head()
```

```python
Special1617 = pd.merge(Special16, Special17, how='left', on=['specialty_description']).reset_index(drop = True)
Special1617 = Special1617.dropna()
Special1617['Decrease_in_Ratio'] = (Special1617['npi16'] - Special1617['npi17'])/Special1617['npi17']
Special1617['Decrease_in_Ratio'].max()
```
